### Clean DataFrames

##### US State Energy Data Cleaning

```python
# Drop columns added as they were part of excel header but not actual columns
filtered_State_Energy_Data2018=State_Energy_Data2018.drop(columns=['Note: Rankings are based on the full source data values.','Federal offshore production is not included in the Production Shares.'])
```

```python
# Renaming columns to meningfyl and shorter name
filtered_State_Energy_Data2018.columns=["State","Production Share","Production Rank","Consumption per capita","Consumption Rank"
                                       ,"Expenditure per capita","Expenditure Rank","Carbon Dioxide Emission Rank","Carbon Dioxide Emission"]
```

```python
#Checking fitered data
filtered_State_Energy_Data2018.head()
```

##### US Annual Renewable Energy Data Cleaning

```python
#Replace Not Available values to nan for few columns
cols=["Biofuels Production","Geothermal Energy Consumption","Solar Energy Consumption","Wind Energy Consumption"
    ,"Waste Energy Consumption","Biofuels Consumption"]
US_annual_renewable[cols]
US_annual_renewable[cols]=US_annual_renewable[cols].astype(str).replace("Not Available",np.nan).astype(float)
US_annual_renewable.head()
```

```python
#Renaming column namecorrectly to Year
US_annual_renewable=US_annual_renewable.rename(columns={"Annual Total(trillion Btu)":"Year"})
US_annual_renewable.head()
```

##### US Yearly Energy Production Data Per Source Cleaning

```python
#Dropping drop first row
Filtered_US_Yearly_Energy_Production_by_Source=US_Yearly_Energy_Production_by_Source.drop(0)
```

```python
Filtered_US_Yearly_Energy_Production_by_Source.head()
```

```python
#Replace Not Available values to nan for few columns
cols=["Geothermal Energy Production","Solar Energy Production","Wind Energy Production"]

Filtered_US_Yearly_Energy_Production_by_Source[cols]=Filtered_US_Yearly_Energy_Production_by_Source[cols].astype(str).replace("Not Available",np.nan).astype(float)
Filtered_US_Yearly_Energy_Production_by_Source.head()
```

```python
#Renaming column appropriately to year
Filtered_US_Yearly_Energy_Production_by_Source=Filtered_US_Yearly_Energy_Production_by_Source.rename(columns={"Annual Total((Quadrillion Btu))":"Year"})

#Converting Year column to Int to remove .0 from value
Filtered_US_Yearly_Energy_Production_by_Source["Year"]=Filtered_US_Yearly_Energy_Production_by_Source["Year"].astype(int)
```

```python
Filtered_US_Yearly_Energy_Production_by_Source.head()
```

##### US Monthly Renewable data Cleaning

```python
US_monthly_renewable.head()
```

```python
#Replace Not Available values to nan for few columns
cols=["Wood Energy Production","Biofuels Production","Total Biomass Energy Production","Total Renewable Energy Production" ,"Hydroelectric Power Consumption",
      "Geothermal Energy Consumption","Solar Energy Consumption","Wind Energy Consumption" ,"Wood Energy Consumption"
      ,"Waste Energy Consumption","Biofuels Consumption","Total Biomass Energy Consumption" ,"Total Renewable Energy Consumption"]
#US_monthly_renewable[cols]
US_monthly_renewable[cols]=US_monthly_renewable[cols].astype(str).replace("Not Available",np.nan).astype(float)
US_monthly_renewable.head()
```

```python
#Rename column to Year/Month
US_monthly_renewable=US_monthly_renewable.rename(columns={"Month(Trillion Btu)":"Year/Month"})
US_monthly_renewable.head()
```

```python
#Split Year and Month
US_monthly_renewable[['Year','Month']] = US_monthly_renewable["Year/Month"].apply( 
   lambda x: pd.Series(str(x).split(" ")))

US_monthly_renewable.head()
```

### Add dataframes into Mongo db database

```python
#create database connection with mongodb
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

#Create US_Eneregy_db database
db = myclient["US_Energy_db"]

```

##### Load dataframes to database

```python
##Create and Load State_EnergyData2018 collection in mongodb from filtered_State_Energy_Data2018 dataframe

collection=db['State_EnergyData2018']

filtered_State_Energy_Data2018_dict = filtered_State_Energy_Data2018.to_dict("records")

#insert collection
collection.insert_many(filtered_State_Energy_Data2018_dict)
```

```python
##Create and Load US_annual_renewable collection in mongodb from US_annual_renewable dataframe

collection=db['US_annual_renewable']
#US_annual_renewable.reset_index(inplace=True) #.reset_index(level=0, drop=True)#
US_annual_renewable_dict = US_annual_renewable.to_dict("records")
#insert collection
collection.insert_many(US_annual_renewable_dict)
```

```python
##Create and Load US_monthly_renewable collection in mongodb from US_monthly_renewable dataframe

collection=db['US_monthly_renewable']
#US_monthly_renewable.reset_index(level=0, drop=True)#.reset_index(inplace=True)
US_monthly_renewable_dict = US_monthly_renewable.to_dict("records")
#insert collection
collection.insert_many(US_monthly_renewable_dict)
```

```python
##Create and Load US_Yearly_Energy_Production_by_Source collection in mongodb from 
##Filtered_US_Yearly_Energy_Production_by_Source dataframe

collection=db['US_Yearly_Energy_Production_by_Source']
#US_Yearly_Energy_Production_by_Source.reset_index(level=0, drop=True)#.reset_index(inplace=True)
US_Yearly_Energy_Production_by_Source_dict = Filtered_US_Yearly_Energy_Production_by_Source.to_dict("records")
#insert collection
collection.insert_many(US_Yearly_Energy_Production_by_Source_dict)
```
